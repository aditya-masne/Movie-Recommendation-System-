```python
import ast
```

```python
def keyname(obj):
    y=[]
    for i in ast.literal_eval(obj):
        name=i['name']
        y.append(name)
    return y    
```

```python
dataset['genres']=dataset['genres'].apply(keyname)
```

```python
dataset.keywords.iloc[0]
```

```python
def keyname2(obj2):
    y2=[]
    for i in ast.literal_eval(obj2):
        name2=i['name']
        y2.append(name2)
    return y2    
```

```python
dataset['keywords']=dataset['keywords'].apply(keyname2)
```

```python
dataset['overview']=dataset['overview'].apply(lambda x:x.split())
```

```python
dataset['overview'][0]
```

```python
dataset['keywords']=dataset['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
dataset['genres']=dataset['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
dataset['tags']=dataset['genres']+dataset['keywords']+dataset['overview']
```

```python
dataset
```

```python
dataset=dataset.drop(['keywords','overview','genres'],axis=1)	
```

```python
dataset
```

```python
dataset['tags']=dataset['tags'].apply(lambda x:" ".join(x))
```

```python
dataset['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(dataset['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommand(movie):
    movie_index_similarity=dataset[dataset['title']==movie].index[0]
    movie_index_title=(sorted(list(enumerate(similarity[movie_index_similarity])),reverse=True,key=lambda x: x[1])[1:6])
    for i in movie_index_title:
        name=dataset.iloc[i[0]].title
        print(name)
        # print(i)
```

```python
recommand('Avatar')
```
